# **Project Summary -**

The Restaurant Analytics & Sentiment Analysis using Machine Learning and NLP project aims to analyze restaurant information and customer reviews to generate meaningful business insights and automatically identify customer sentiment. The project combines Exploratory Data Analysis (EDA), Natural Language Processing (NLP), Machine Learning Classification, and Unsupervised Learning (Clustering) to understand customer behavior and restaurant performance.

The project utilizes two datasets: a Restaurant dataset containing restaurant-related information such as cost, cuisines, collections, timings, and ratings, and a Customer Reviews dataset containing customer reviews, review ratings, pictures, and reviewer details. The first step involved data preprocessing, where missing values, duplicate records, and incorrect data types were handled to improve data quality. Outliers in numerical features were identified using boxplots and treated using the Interquartile Range (IQR) method wherever required. Additional features such as Cost Category, Cuisine Count, Review Length, and Character Count were created to enhance model performance.

Extensive Exploratory Data Analysis (EDA) was performed using histograms, bar charts, scatter plots, boxplots, heatmaps, pair plots, and correlation analysis. The analysis revealed several business insights, including the distribution of restaurant costs, customer ratings, popular cuisines, restaurant collections, customer-uploaded pictures, and the relationship between restaurant cost and customer ratings. Statistical hypothesis testing using Pearson Correlation was conducted to validate relationships between important numerical variables.

For the NLP pipeline, customer review text underwent several preprocessing steps, including expanding contractions, converting text to lowercase, removing punctuation, removing URLs and digits, removing stopwords, tokenization, lemmatization, part-of-speech tagging, and TF-IDF vectorization. Customer ratings were converted into three sentiment classes (Positive, Neutral, and Negative) to create the target variable for sentiment classification.

Three supervised machine learning models were developed for sentiment prediction: Logistic Regression, Multinomial Naive Bayes, and Random Forest Classifier. The dataset was divided into training and testing sets using an 80:20 split, and model performance was evaluated using Accuracy, Precision, Recall, and F1-Score. Hyperparameter tuning was performed using GridSearchCV to improve model performance and identify the optimal model configuration. The performance of all three models was compared, and the best-performing model was selected based on the evaluation metrics.

In addition to sentiment classification, K-Means Clustering was applied to the restaurant dataset to group restaurants based on business characteristics such as cost, cuisine diversity, collections, and pricing category. This clustering approach helps identify similar restaurant segments, supporting market analysis and business decision-making.

The project demonstrates how Machine Learning and Natural Language Processing can be effectively applied to restaurant analytics. By automatically classifying customer reviews into positive, neutral, and negative sentiments, businesses can monitor customer satisfaction, identify common issues, and improve service quality. The clustering results further assist in understanding restaurant categories and customer preferences, enabling data-driven business strategies. Overall, this project provides a practical solution for restaurant performance analysis, customer feedback analysis, and intelligent decision-making using data analytics and machine learning.

## Step 3: Missing Value Treatment

### Objective

The objective of this step is to handle missing values present in the datasets and improve overall data quality.

Missing values can negatively impact:
- Data analysis
- Statistical calculations
- Machine learning models
- Business insights

Therefore, appropriate treatment strategies are applied based on the importance and nature of each variable.

### Missing Value Treatment Strategy

#### Restaurant Metadata Dataset
- Missing values in `Collections` will be replaced with "Unknown".
- Missing values in `Timings` will be replaced using the most frequently occurring value (Mode).

#### Restaurant Reviews Dataset
- Records with missing review text will be removed because review text is essential for sentiment analysis.
- Missing values in `Reviewer`, `Metadata`, and `Time` will be replaced with "Unknown".
- Missing ratings will be retained for now and handled separately during model preparation if required.

````python
# =============================================================================
# Step 3: Missing Value Treatment
# =============================================================================

print("Missing Values Before Treatment")
print("=" * 60)

print("\nRestaurant Metadata Dataset")
print(restaurant_clean.isnull().sum())

print("\nRestaurant Reviews Dataset")
print(reviews_clean.isnull().sum())


# ------------------------------------------------------------------
# Restaurant Metadata Dataset
# ------------------------------------------------------------------

# Fill missing collections with Unknown
restaurant_clean["Collections"] = restaurant_clean["Collections"].fillna("Unknown")

# Fill missing timings with mode
restaurant_clean["Timings"] = restaurant_clean["Timings"].fillna(
    restaurant_clean["Timings"].mode()[0]
)


# ------------------------------------------------------------------
# Restaurant Reviews Dataset
# ------------------------------------------------------------------

# Remove rows with missing review text
reviews_clean = reviews_clean.dropna(subset=["Review"])

# Fill categorical missing values
reviews_clean["Reviewer"] = reviews_clean["Reviewer"].fillna("Unknown")
reviews_clean["Metadata"] = reviews_clean["Metadata"].fillna("Unknown")
reviews_clean["Time"] = reviews_clean["Time"].fillna("Unknown")


# ------------------------------------------------------------------
# Missing Values After Treatment
# ------------------------------------------------------------------

print("\n")
print("=" * 60)
print("Missing Values After Treatment")
print("=" * 60)

print("\nRestaurant Metadata Dataset")
print(restaurant_clean.isnull().sum())

print("\nRestaurant Reviews Dataset")
print(reviews_clean.isnull().sum())

print("\nRestaurant Metadata Dataset Shape:", restaurant_clean.shape)
print("Restaurant Reviews Dataset Shape:", reviews_clean.shape)
````

### Observation

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

### Null Hypothesis (H₀)

There is no significant difference in customer ratings across different restaurant cost categories (Budget, Mid-Range, Premium, and Luxury).

### Alternate Hypothesis (H₁)

There is a significant difference in customer ratings across different restaurant cost categories (Budget, Mid-Range, Premium, and Luxury).

#### 2. Perform an appropriate statistical test.

````python
from scipy.stats import f_oneway

# Create cost categories if not already created

merged_df["Cost_Category"] = pd.cut(
    merged_df["Cost"],
    bins=[0, 500, 1000, 1500, 3000],
    labels=["Budget", "Mid-Range", "Premium", "Luxury"]
)

# Extract ratings for each group

budget = merged_df[merged_df["Cost_Category"]=="Budget"]["Rating"]
midrange = merged_df[merged_df["Cost_Category"]=="Mid-Range"]["Rating"]
premium = merged_df[merged_df["Cost_Category"]=="Premium"]["Rating"]
luxury = merged_df[merged_df["Cost_Category"]=="Luxury"]["Rating"]

# ANOVA Test

f_stat, p_value = f_oneway(
    budget,
    midrange,
    premium,
    luxury
)

print("F-Statistic:", round(f_stat,4))
print("P-Value:", p_value)

if p_value < 0.05:
    print("\nReject Null Hypothesis (H₀)")
else:
    print("\nFail to Reject Null Hypothesis (H₀)")
````

##### Which statistical test have you done to obtain P-Value?

One-Way ANOVA Test (Analysis of Variance)



##### Why did you choose the specific statistical test?

One-Way ANOVA was selected because the analysis involves comparing the mean customer ratings across more than two independent groups (Budget, Mid-Range, Premium, and Luxury restaurants). ANOVA helps determine whether the differences observed among the group means are statistically significant and not due to random variation.

StandardScaler transforms numerical features by centering them around a mean of 0 and scaling them to have a standard deviation of 1. This ensures that features with different ranges contribute equally during model training.

- Leads to faster convergence and more stable machine learning models.

### 8. Data Splitting

````python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF Vectorization
tfidf = TfidfVectorizer(max_features=5000)

X = tfidf.fit_transform(reviews_clean["Review"])

# Target Variable
y = reviews_clean["Sentiment"]

print("TF-IDF Shape:", X.shape)
print("Target Shape:", y.shape)


from sklearn.model_selection import train_test_split

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Data Shape :", X_train.shape)
print("Testing Data Shape  :", X_test.shape)

print("\nTraining Labels :", y_train.shape)
print("Testing Labels  :", y_test.shape)
````

##### What data splitting ratio have you used and why?

### Data Splitting Ratio

An **80:20 train-test split** was used.

- **80%** of the data was used for training the machine learning models.
- **20%** of the data was reserved for testing and evaluating model performance on unseen data.

### Why 80:20?

- It provides sufficient data for the model to learn meaningful patterns.
- It retains enough unseen data for reliable evaluation.
- A fixed `random_state=42` was used to ensure reproducibility.
- `stratify=y` was applied to preserve the class distribution (Positive, Neutral, Negative) in both the training and testing datasets, preventing class imbalance during model evaluation.

## ***7. ML Model Implementation***

### ML Model - 1

````python
# =====================================================
# ML Model 1 : Logistic Regression
# =====================================================

from sklearn.linear_model import LogisticRegression

# Create Model
lr_model = LogisticRegression(max_iter=1000, random_state=42)

# Train Model
lr_model.fit(X_train, y_train)

print("Logistic Regression Model Trained Successfully.")

## Prediction

y_pred_lr = lr_model.predict(X_test)

print("Prediction Completed.")


from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix
)

accuracy = accuracy_score(y_test, y_pred_lr)
precision = precision_score(y_test, y_pred_lr, average='weighted')
recall = recall_score(y_test, y_pred_lr, average='weighted')
f1 = f1_score(y_test, y_pred_lr, average='weighted')

print("Accuracy :", round(accuracy,4))
print("Precision:", round(precision,4))
print("Recall   :", round(recall,4))
print("F1 Score :", round(f1,4))

print("\nClassification Report\n")
print(classification_report(y_test, y_pred_lr))



import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_test, y_pred_lr)

plt.figure(figsize=(6,5))
sns.heatmap(cm,
            annot=True,
            fmt='d',
            cmap='Blues',
            xticklabels=lr_model.classes_,
            yticklabels=lr_model.classes_)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Logistic Regression")

plt.show()
````

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

````python
metrics = {
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1 Score": f1
}

plt.figure(figsize=(7,5))

plt.bar(metrics.keys(), metrics.values())

plt.ylim(0,1)

plt.title("Logistic Regression Evaluation Metrics")

plt.ylabel("Score")

plt.show()
````

Logistic Regression was used as the first machine learning model for sentiment classification. It is a supervised learning algorithm that predicts the sentiment class (Positive, Neutral, or Negative) based on the TF-IDF features extracted from customer reviews.

The model was evaluated using Accuracy, Precision, Recall, and F1-Score. The evaluation metric score chart shows that the model achieved balanced performance across all metrics, with scores around 80–83%. This indicates that the model can classify customer sentiments effectively while maintaining a good balance between precision and recall.

Overall, Logistic Regression provides a strong baseline model for sentiment analysis due to its simplicity, efficiency, and good performance on text classification problems.

#### 2. Cross- Validation & Hyperparameter Tuning
